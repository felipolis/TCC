```python
import librosa

y, sr = librosa.load("./data/train_audio/abethr1/XC128013.ogg", sr=None)

print(sr)
```

```python
(1/32000) * 2048
```

```python
import numpy as np

D = librosa.amplitude_to_db(np.abs(librosa.stft(y, n_fft=2048, hop_length=1024)), ref=np.max)
```

```python
print(D.shape)
```

```python
45 / 0.064 * 2
```

```python
from matplotlib import pyplot as plt

fig = plt.figure(figsize=(1426/72, 1025/72), dpi=72)
ax = plt.Axes(fig, [0., 0., 1., 1.])
ax.set_axis_off()
fig.add_axes(ax)
plt.pcolormesh(D)
plt.savefig("./XC128013.png", dpi=72, pad_inches=0, bbox_inches='tight')
```
